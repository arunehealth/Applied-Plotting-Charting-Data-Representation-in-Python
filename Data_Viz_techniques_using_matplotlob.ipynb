```
#animation
import matplotlib.animation as animation
```

```
#creating a random data array
n = 100
x = np.random.randn(n)
print (x)
```

**Define a cut-off for animation.** 
Creating a function to do the plotting.
Function is called update. 

matplotlib FuncAnimation object is going to call this function  
every few milliseconds and pass in the frame number we are on starting 
with frame zero. 

So we can use this is as the index into our array values,
which we called x. 


curr is current frame of the entire data frame

a.event is the animation event

 a.event_source.stop() asks the function update to stop if the current frame is at the end of the frame
 
 To set and evenly space bins - use the NumPy arange function. 
 
 To ensure that the bins don't change - use the balance of minus 4 to plus 4, in half-step increments. 

```

def update(curr):
  if curr == n:
    a.event_source.stop()
    plt.cla()
  bins = np.arrange(-4,4,bins=0.5)
  plt.hist(x[:curr], bins=bins)
  plt.axis(-4,4,0,30)
  plt.gca().set_title('NormalDist')
  plt.gca().set_ylabel("Freq")
  plt.gca().set_xlabel("Value")
  #text function called annotate- places text at a certain position in the 
  #chart and shows how many samples are currently being rendered to the screen. 
  plt.annotate('n={}'.format(curr),[3,27])
  
```

```
#first parameter is the figure that we're working with. 
#hen the name of function
#Then amount of time between updates. 
#Set to variable a. Ifnot, function do not  know how to stop  animation. 
fig = plt.figure()
a=animation.FuncAnimation(fig, update, interval = 1000)
```
